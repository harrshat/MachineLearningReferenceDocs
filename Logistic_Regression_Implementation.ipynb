```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
# the data generated is standardized so no need for Standard scaler,etc.
```

```python
X.shape
```

```python
y.shape
```

```python
df=pd.DataFrame(X)
```

```python
df.head()
```

```python
df['Output']=y
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
#class_weight can be used in case of imbalanced dataset
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
print(y_pred)
```

```python
# For probability estimates --The returned estimates for all classes are ordered by the label of classes.
logistic.predict_proba(X_test)
# eg.1 [7.74477909e-01, 2.25522091e-01] --[0 1]-- since value in place of 0 is higher, output - 0
```

```python
# Performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
#classification_report for precision and recall

score=accuracy_score(y_test,y_pred)
print("Score: ",score,"\n")
print("__________________________________________________________________\n")
conf_matr=confusion_matrix(y_test,y_pred)
print("Confusion Matrix: \n",conf_matr)
print("__________________________________________________________________\n")
class_report=classification_report(y_test,y_pred)
print("Classification Report: \n",class_report)
```
